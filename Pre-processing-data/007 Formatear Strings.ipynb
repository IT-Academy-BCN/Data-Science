Diferentes formas de hacer que una cadena de literales ``string`` se alarge en mรบltiples lรญneas :

- Insertamos el <span style=color:red> \n (barra diagonal invertida **n**)</span> entre los nombres de los meses. ``\n`` genera, en ese punto, un *nuevo carรกcter de lรญnea* en la cadena.
- El uso del carรกcter ``\ (barra diagonal invertida)`` es una forma de tratar con caracteres difรญciles de escribir en una cadena, como **secuencias de escape / escape sequenceses**.
- Otra secuencia de escape es escapar de ``comillas simples ''`` o ``comillas dobles ""``.
- ``Comillas triples "" "" `` funciona como una cadena, y puede incluir tantas lรญneas de texto hasta que cierre la cadena escribiendo nuevamente `` """``.

```python
dias = "Lun Mar Mie Jue Vie Sab Dom"
meses = "\nEne \nFeb \nMar \nAbr \nMay \nJun \nJul \nAug"

print("Aqui estan los dias: ", dias)
print(" y aquรญ los meses: ", meses)

print("""
Algo estรก sucediendo aquรญ.
Con las tres comillas dobles.
Podremos escribir todo lo que queramos.
Incluso 4 lรญneas si queremos, o 5, o 6.
""")
```

```python
print("Yo mido 6'2\" de alto.") # escapamos doble-comillas dentro del string
print('Yo mido 6\'2" de alto.') # escapamos comillas individuales dentro del string
```
