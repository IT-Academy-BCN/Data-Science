# <span style=color:darkblue>INTRODUCTION TO DATA</span>

Statistics is the study of how best to **collect**, **analyze**, and **draw conclusions** from **``data``**.

Statistics is a general *<u>process of investigation</u>* :

<span style=color:darkblue><em><b>1. Identify a question or problem.</b></em>   
<span style=color:darkblue><em><b>2. Collect relevant data on the topic.</b></em>   
<span style=color:darkblue><em><b>3. Analyze the data.</b></em>   
<span style=color:darkblue><em><b>4. Form a conclusion.</b></em>   

Statistics focuses on making stages <b>2-4</b> <span style=color:blue>objective, rigorous, and efficient</span>.   
   
Statistics has <span style=color:blue><b>three</b></span> primary components : 

- How best can we <span style=color:blue>collect</span> data?
- How should it be <span style=color:blue>analyzed</span>?
- And what can we <span style=color:blue>infer</span> from the analysis?

## <span style=color:green>TYPES OF VARIABLES</span>

Generally, in Math and Statistics <span style=color:blue>variables</span> may be <span style=color:blue>Numerical</span> or <span style=color:blue>Categorical Variables</span>.

A <span style=color:blue>Variable</span> is a <u>quantity whose value changes</u>. 

### <span style=color:darkred>NUMERICAL</span>

A <span style=color:blue>Numerical variable</span> can take a wide range of numerical values, and it is sensible to <span style=color:blue><em>add, subtract</em></span>, or take <span style=color:blue><em>averages</em></span> with those <span style=color:blue>values</span>. On the other hand, we would <b><u>not</u></b> classify a variable reporting <em>"<u>telephone area codes</u>"</em> as <span style=color:blue>numerical</span> since there is **no** sense to *average, sum*, and *difference*.

#### <span style=color:gray><u>Discrete Variables</u></span>

A <span style=color:blue><b><u>Discrete variable</u></b></span> is a <span style=color:blue>variable whose value is obtained by <u>counting</u>.</span> 

Over a particular range of real values ($\:\mathbb {R}\:$) is any value in the range that the variable is permitted to take on, there is a positive minimum distance to the nearest other permissible value. The number of permitted values is either <span style=color:blue><b><em><u>finite or countably infinite</u></em></b>. 

Common examples are variables that must be <span style=color:blue><em>integers, non-negative integers, positive integers</em></span>, or <span style=color:blue><em>only the integers 0 and 1</em></span>.

<u><b>Examples:</b></u> 
- number of students present
- number of red marbles in a jar
- number of heads when flipping three coins
- students‚Äô grade level

#### <span style=color:gray><u>Continuous Variables</u></span>

A <span style=color:blue><b><u>Continuous variable</u></b></span> is a <span style=color:blue>variable whose value is obtained by measuring.</span>

A <span style=color:blue><b><u>Continuous variable</u></b></span> is one which can take on <span style=color:blue>infinitely many, uncountable values</span>.

For example, a variable over a non-empty range of the real numbers ($\:\mathbb {R}\:$) ***a*** and ***b*** is <span style=color:blue><b><u>continuous</u></b></span>, if it can take on <span style=color:blue>*any value in that range*</span>. The reason is that any range of real numbers between ***a*** and ***b*** with$\hspace{3mm}$${\displaystyle a,b\in \mathbb {R} ;   
\hspace{4mm}a\neq b}$ is infinite and uncountable.

<u><b>Examples:</b></u> 
- height of students in class
- weight of students in class
- time it takes to get to school
- distance traveled between classes

### <span style=color:darkred>CATEGORICAL</span>

A <span style=color:blue><b><u>Categorical Variable</u></b></span> is one that has two or more categories.   
   
A <span style=color:blue><b><u>Categorical Variable</u></b></span> takes on a limited, and usually fixed, number of possible values, categories;  and the possible values are call the <span style=color:blue><em><u>variable's <b>levels</b>.</u></em></span>

Types of <span style=color:blue>Categorical Variables</span> : 

<span style=color:blue>1.- <b>Nominal</b> Variables</span>, where their <span style=color:blue>levels</span> have a natural order and are <span style=color:blue><ins> deduced from the data </ins> </span>.

# <SPAN style=color:darkblue>OVERVIEW OF DATA COLLECTION PRINCIPLES</SPAN>

The <span style=color:blue><b>first step</b></span> in conducting research is to <span style=color:blue>identify topics or questions that are to be investigated</span>.   

A <span style=color:blue>clearly laid out research question</span> is helpful in <span style=color:blue><b>identifying</b></span> what <span style=color:blue><u>subjects</u></span> or <span style=color:blue><u>cases</u> should be <b>studied</b></span> and what <span style=color:blue>variables are <b>important</b></span>.   

It is also important to <span style=color:blue>consider how data are collected</span> so that they <span style=color:blue><b>are reliable</b></span> and help <span style=color:blue><b>achieve</b></span> the <span style=color:blue><u>research goals</u></span>.

## <SPAN style=color:green>Populations and Samples</SPAN>

Consider the following three research questions:

1. What is the average mercury content in swordfish in the Atlantic Ocean?
2. Over the last 5 years, what is the average time to complete a degree for Duke undergraduate students?
3. Does a new drug reduce the number of deaths in patients with severe heart disease?

Each research question refers to a <span style=color:blue><b><u>target population</u></b></span>, and each observation a <span style=color:blue><b>case</b></span>. Generally, it is too expensive to collect data for every <span style=color:blue><b>case in a population</b></span>. Instead, a <span style=color:blue><b>Sample</b></span> is taken. 

A <span style=color:blue><b>Sample</b></span> represents a **subset of the cases** and is often a **small fraction of the population**. For instance, 60 swordfish in the population might be selected, and this <span style=color:blue><b>sample data</b></span> may be used to provide an <span style=color:blue><b>estimate of the population average</b></span> and <span style=color:blue><b>answer the research question</b>.

<b><u>Example :</u></b>   

*For the three research questions above, identify the <span style=color:blue><b><u>target population</u></b></span> and what represents an individual <span style=color:blue><b><u>case</u></b></span>.*

* (1) In the first question, the target population is all swordfish in the Atlantic ocean, and each fish represents a <span style=color:blue><b>case</b>   
$%$
* (2) Notice that the first question is only relevant to students who complete their degree; the average cannot be computed using a student who never finished her degree. Thus, only Duke undergraduate students who have graduated in the last five years represent <span style=color:blue>cases in the population</span> under consideration.  Each such student would represent an individual <span style=color:blue><b>case</b>.   
$%$   
* (3) The population includes all people with severe heart disease. A person with severe heart disease represents a <span style=color:blue><b>case</b>. 

## <span style=color:darkgreen>Anecdotal Evidence</span>

Consider the following possible responses to the three research questions:

1. A man on the news got mercury poisoning from eating swordfish, so the average mercury concentration in swordfish must be dangerously high.
2. I met two students who took more than 7 years to graduate from Duke, so it must take longer to graduate at Duke than at many other colleges.
3. My friend‚Äôs dad had a heart attack and died after they gave him a new heart disease drug, so the drug must not work.

Each conclusion is based on data. However, there are two problems : 

- First, the data only represent one or two cases. 
- Second, and more importantly, it is unclear whether these cases are actually representative of the population. 

Data collected with this lack of criteria are called <span style=color:blue><b><u>anecdotal evidence</u></b></span>.

*Instead of looking at the most unusual cases, <span style=color:blue>we should examine a sample of <u>many cases</u> that represent the population.</span>*

## <span style=color:darkgreen>Sampling from a population</span>

<span style=color:blue><b>Sampling</b></span> is a method that allows researchers to <span style=color:blue><u>infer</u></span> information about a <span style=color:blue><b>population</b></span> based on results from a <span style=color:blue>subset of the population</span>, without having to investigate every individual.

<b>Reducing the number</b> of individuals in a study reduces the <b>cost and workload</b>, and may make it easier to obtain high quality information, but this has to be <span style=color:blue><b>balanced</b></span> <span style=color:red>against</span> having a <span style=color:blue>large enough <b>sample size</b></span> with enough <span style=color:blue>power to detect a <b>true association</b></span>.

### <span style=color:darkred><u>Bias/Parcialidad in sampling</u></span>

It is important that the <span style=color:blue>individuals</span> selected are <span style=color:blue>representative</span> of the whole <span style=color:blue>population</span>.
This may involve specifically targeting <span style=color:red><b>hard to reach groups</b></span>.   

For example, if the electoral roll for a town was used to identify participants, some people, such as the <span style=color:red>homeless</span>, would not be registered and therefore excluded from the study by default.

*Why pick a sample randomly?$\hspace{5mm}$Why not just pick a sample by hand?*

<b><u>Example 1</u></b>   

<i>Suppose we ask a <b>student</b> who happens to be <b>majoring in nutrition</b> to select several graduates for the study. 

- What kind of students do you think she might collect?   
$%$   
- Do you think her sample would be representative of all graduates?</i>

Perhaps she would pick a disproportionate number of graduates from <span style=color:blue>health-related fields</span>. Or perhaps her selection would be <span style=color:blue>well-representative</span> of the population. 

If someone was permitted to pick and choose exactly which graduates were included in the <span style=color:blue>sample</span>, it is entirely possible that the sample could be <span style=color:blue>skewed</span> to that person‚Äôs interests, which may be entirely <span style=color:blue>unintentional</span>.

When selecting samples by hand, we run the risk of picking a <span style=color:blue>biased sample - <u>bias into a sample</u></span>, even if that bias is <span style=color:blue>unintentional</span> or <span style=color:blue>difficult to discern</span>.

<span style=color:blue>Sampling randomly</span> helps resolve this problem.

There are <span style=color:blue>five important potential sources of bias</span> that should be considered when selecting a <span style=color:blue>sample</span>, irrespective of the method used. 

Sampling bias may be introduced when:

1. Any <span style=color:blue>pre-agreed sampling rules</span> are deviated from
2. People in <span style=color:blue>hard-to-reach groups</span> are omitted
3. <span style=color:blue>Selected individuals are replaced</span> with others, for example if they are difficult to contact
4. There are <span style=color:blue>low response rates</span>
5. An <span style=color:blue>out-of-date list</span> is used as the sample frame (for example, if it excludes people who have recently moved to an area).

<div class="alert alert-warning alert-warning"><span style=color:darkred><b><em>SIMPLE RANDOM SAMPLE</em></b></span>

The most **basic random sample** is called a <span style=color:blue><b><u>Simple Random Sample</u></b></span>, each case in the population has an equal chance of being included and there is no implied connection between the cases in the sample. Taking <span style=color:blue><b><u>Simple Random Sample</u></b></span> helps minimize <span style=color:blue><b><u>bias</u></b></span>.</div>

Although <span style=color:blue><b><u>Simple Random Sample</u></b></span> helps minimize bias, bias can crop up in other ways. Even when cases are picked up at random, caution must be exercised if the <span style=color:blue>non-response bias</span> is high, and can skew results. Another common downfall is a <span style=color:blue>convenience sample</span>, where individuals who are easily accessible are more likely to be included in the sample.   
   
$%$   
<center><span style=color:blue>It is often difficult to discern what sub-population a convenience sample represents</span></center>.

### <span style=color:darkred>Explanatory and response variables</span>

Consider the following question from $county$ dataset :

- ***Is federal spending, on average, higher or lower in counties with high rates of porverty ?***

If we suspect poverty might affect spending in a *county*, then poverty is the <span style=color:blue>explanatory/explicativa variable</span> and federal spending is the <span style=color:blue>response variable</span> in the relationship. If there are many variables, it may be possible to consider a number of them as <span style=color:blue>explanatory variables</span>. 

Sometimes the <span style=color:blue>explanatory variable</span> is called the <span style=color:blue>independent variable</span> and the <span style=color:blue>response variable</span> is called the <span style=color:blue>dependent variable</span>. However, this becomes confusing since a pair of variables might be <span style=color:blue>independent</span> or <span style=color:blue>dependent</span>, so we avoid this language.

To <span style=color:red><em>identify</em></span> the <span style=color:blue>explanatory variable</span> in a <span style=color:blue><u>pair of variables</u></span>, <span style=color:red><em>identify</em></span> which of the two is <span style=color:blue><b>suspected of <u>affecting</u></b></span> the other and plan an appropiate analysis.

<div class="alert alert-warning alert-warning"><span style=color:darkred><b><em>ASSOCIATION DOES NOT IMPLY CAUSATIOM</em></b></span>

Labeling variables as <b>explanatory</b> and <b>response</b> does <b>not guarantee</b> the relationship between the two is actually <b>causal</b>, even if there is an association identified between the two variables. <b><i>We use these labels only to keep track of which variable we suspect affects the other</i></b>. Even in some cases, there is <b>no explanatory</b> or <b>response variable</b>.</div>

In general, <span style=color:red>association</span> does not imply <span style=color:red>causation</span>,  and <span style=color:red>causation</span> can only be inferred from a <span style=color:red><b>randomized experiment</b></span>.

### <SPAN style=color:darkred>Introducing Observational Studies and Experiments</SPAN>

There are <span style=color:blue>two primary types</span> of data collection : <span style=color:blue>Observational Studies</span> and <span style=color:blue>Experiments</span>.

- <span style=color:blue><u>Observational Studies</u></span> when data is collected in a way that does <b>not</b> directly <b>interfere</b> with how the data arise. <span style=color:blue>We merely observe the data that arise</span>.   
$%$   
- <span style=color:blue><u>Observational Studies</u></span> can provide <i><u>evidence</u></i> of naturally <i><u>ocurring association between variables</u></i>, but they cannot by themselves show a <b><u>causal connection</u></b>.   

When we want to investigate the possibility of a <span style=color:blue>causal connection</span>, we will conduct an <span style=color:blue><u>Experiment</u></span>. And ususally there will be both an <b>explanatory</b> and a <b>response variable</b>.

When individuals are <span style=color:blue>randomly assigned</span> to a group, the experiment is called a <span style=color:blue>randomized experiment</span>.

# <SPAN style=color:darkblue>OBSERVATIONAL STUDIES</SPAN>

Generally, data in <span style=color:blue><u>Observational Studies</u></span> are collected only by monitoring what occurs, while <span style=color:blue><u>experiments</u></span> require the <b>primary explanatory variable</b> in a study be <b>assigned</b> to each subject by the researchers.

Making causal conclusions based on <span style=color:blue><u>Experiment</u></span> is often reasonable</u>. However, making the same <i>causal conclusions</i> based on <b>observational data</b> can be misleading and is not recommended. Thus, <b>observational studies</b> are generally <em>only sufficient to show associations.</em>

<span style=color:blue><b><u>Confounding variable</u></b></span> is a variable that is correlated with both the <b>explanatory</b> and <b>response variable</b>. However, there is no guarantee that all <span style=color:blue>confounding variables</span> can be examined or measured.   

The $county$ dataset is an <span style=color:blue>observational</span> study with <span style=color:blue>confounding variables</span>, and its data cannot easily be used to make <span style=color:blue><b>causal conclusions</b></span>.

<span style=color:blue>Observational studies</span> come in two forms: 

1. <span style=color:blue><b>Prospective studies</b></span> identifies individuals and collects information as events unfold.
2. <span style=color:blue><b>Retrospective studies</b></span> collect data after events have taken place.

Some datasets may contain both prospectively and retrospectively-collected variable, such as *county datasets*, Local governments prospectively collect some variables as events unfolded (e.g. retails sales) while the federal government retrospectively collected others during the 2010 census (e.g. county population counts).

# <span style=color:darkblue>EXPERIMENTS</span>

Studies where the researchers assign treatments to cases are called experiments. 

When this assignment includes randomization, e.g. using a coin flip to decide which treatment a patient receives, it is called a <span style=color:blue>Randomized experiment</span>. <span style=color:blue>Randomized experiments</span> are fundamentally important when trying to show a <span style=color:blue><b><u>causal connection between two variables.</u></b></span>

## <span style=color:darkgreen>Principles of Experimental Design</span>

<span style=color:blue>Randomized experiments</span> are generally built on <b>four principles</b> :

1. <span style=color:blue><b><u>Controlling</u></b></span>. Researchers assign treatments to cases, and control any other differences in the groups.  

<b><em>For example</em></b>: 

When patients take a <b>drug in pill form</b>, some patients take the pill with only a <b><i>sip of water</i></b> while others may have it with an entire glass of water. To control for the effect of water consumption, a doctor may ask all patients to drink a 12 ounce glass of water with the pill.

2. <span style=color:blue><b><u>Randomization</u></b></span>. Researchers <b><i>randomize patients into treatment groups</i></b> to account for variables that cannot be controlled.   

<b><em>For example</em></b> : 

Some patients may be more susceptible to a disease than others due to their dietary habits. Randomizing patients into the treatment or control group helps even out such differences, and it also prevents accidental bias from entering the study.

3. <span style=color:blue><b><u>Replication</u></b></span>. The more cases researchers observe, the more accurately they can estimate the effect of the explanatory variable on the response. 

In a single study, we replicate by collecting a sufficiently large sample. Additionally, a group of scientists may replicate an entire study to verify an earlier finding.

4. <span style=color:blue><b><u>Blocking</u></b></span>. Researchers sometimes know or suspect that variables, other than the treatment, influence the response.   

Under these circumstances, they may first group individuals based on this variable into blocks and then randomize cases within each block to the treatment groups. This strategy is often referred to as <span style=color:blue><b><u>blocking</u></b></span>. 

For instance, if we are looking at the effect of a drug on heart attacks, we might first split patients in the study into low-risk and high-risk blocks, then randomly assign half the patients from each block to the control group and the other half to the treatment group. This strategy ensures each treatment group has an equal number of low-risk and high-risk patients.
